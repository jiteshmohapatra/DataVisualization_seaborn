```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1,2,3,4,5,6,7]
y = [4,6,8,16,10,2,12]
```

```python
plt.figure(figsize=(15,6))
plt.plot(x,y,color="green",marker='o',ms=20,mfc='r')
plt.title("Line chart")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [4,6,8,16,10,2,12]

a = [1,2,3,4,5,6,7]
b = [10,12,14,15,16,18,20]

```

```python
plt.figure(figsize=(15,5))
plt.plot(x,y,'g',a,b,'r',marker='o',ms=20,mfc='y',mec='y')
plt.title("Line Graph")
plt.xlabel("X axis")
plt.ylabel("y axis")
plt.show()
```

## bar graph

```python
data1 = ["a","b","c","d","e","f"]
data2 = [98,85,74,69,55,48]
plt.plot(data1,data2,marker='o',ms=20,mfc='y')
plt.title("Bar graph")
plt.xlabel("data1")
plt.ylabel("data2")
plt.legend(loc="upper left", title="List")
plt.show()
```

```python
arr1 = np.array([[1,2,3,4],[4,5,6,7]])
arr2 = np.array([[10,12,13,14],[16,18,20,22]])

```

```python
plt.figure(figsize=(15,6))
plt.plot(arr1,arr2,marker='o',mec='r',mfc='r',ms=10)
plt.title("Barchart")
plt.xlabel("arr1")
plt.ylabel('arr2')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
x = np.array([1,8])
y = np.array([3,10])

plt.plot(x,y,marker='o',ms=20,color="springgreen",mfc='r',mec='y')
plt.title("Line graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
z = np.array([3,8,1,10])
plt.figure(figsize=(10,5))
plt.plot(z,marker='o',ms=20,mfc='deeppink',mec='y',linestyle='dashed',linewidth=5.5,color="orange")
font1 = {"family":"serif","color":"blue","size":20}
font2 = {"family":"verdana","color":"orange","size":15}
plt.title("Line Graph",fontdict=font1,loc='center')
plt.xlabel("X axis",fontdict=font2)
plt.ylabel("Y axis",fontdict=font1)
plt.grid(axis='x',color="green",linestyle="--",linewidth=0.5)
plt.show()
```

## Basic design graph

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
arr1 = np.array([1,2,3,4,5,6,7,8])
arr2 = np.array([10,12,13,14,15,16,17,18])
plt.plot(arr1,arr2,marker='o',linestyle='dashed',linewidth=5.5,color="green",ms=20,mec='r',mfc='y')
font1 = {"family":"serif","color":"orange","size":20}
font2 = {"family":"verdana","color":"blue","size":15}
plt.title("Line Graph",fontdict=font1)
plt.xlabel("x axis",fontdict=font2)
plt.ylabel("y axis",fontdict=font1)
plt.grid(color="cyan",linestyle='--',linewidth='0.5')
plt.show()
```

```python
df1 = pd.read_csv("diabetesdata.csv")
```

```python
plt.plot(df1['Pregnancies'],label='Pregnancies')
plt.plot(df1['Glucose'],label='Glucose')
plt.plot(df1['BloodPressure'],label='BloodPressure')
plt.title("Line chart")
plt.show()
```

```python
plt.hist(df1['Glucose'], bins=10, edgecolor='green',linestyle='dotted')
plt.title('Glucose Distribution')
plt.show()
```

```python
plt.scatter(df1['Age'], df1['BMI'],color='green',linestyle='dashed',linewidth=1.5)
plt.title('Age vs. BMI')
plt.grid(color='orange')
plt.show()

```

```python
plt.barh(df1['Age'], df1['BloodPressure'],color='green',linestyle='dashed',linewidth=1.5)
plt.title('Age vs BloodPressure ')
plt.grid(color='red')
plt.show()
```
