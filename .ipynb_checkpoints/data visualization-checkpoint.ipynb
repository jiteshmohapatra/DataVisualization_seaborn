## Simple Barchart

```python
division = ["A","B","C","D","E"]
marks = [594,561,544,520,493]
plt.bar(division,marks,color="green")
plt.title("bar Graph")
plt.xlabel("division")
plt.ylabel("marks")
plt.show()
```

```python
division = ["A","B","C","D","E"]
marks = [594,561,544,520,493]
plt.barh(division,marks,color="green")
plt.title("bar Graph")
plt.xlabel("division")
plt.ylabel("marks")
plt.show()
```

```python
division = ["A","B","C","D","E"]
marks = [594,561,544,520,493]
plt.plot(division,marks,color="deeppink",marker='s')
plt.title("bar Graph")
plt.xlabel("division")
plt.ylabel("marks")
plt.show()
```

## scatter plot

```python
height = np.array([167,170,152,198,144,99,101,121,175,199,110,152,160,184,179])
weight = np.array([41,56,84,31,99,74,21,64,75,88,50,40,60,71,49])
plt.scatter(height,weight,marker='v',color="red")
plt.title("scatter plot")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

## Pie Chart

```python
forms = ["form A","form B","form C","Form D","Form E"]
marketshare = [20,25,15,10,20]
plt.pie(marketshare,labels=forms,shadow=True,startangle=45)
plt.title("Pie chart")
plt.show()
```

```python
forms = ["form A","form B","form C","Form D","Form E"]
marketshare = [20,25,15,10,20]
plt.pie(marketshare,labels=forms,shadow=True,startangle=45)
plt.title("Pie chart")
plt.legend(loc="upper left",title="List of forms")
plt.show()



forms = ["form A","form B","form C","Form D","Form E"]
marketshare = [20,25,15,10,20]
myexplode = [0.2, 0, 0, 0,0]
plt.pie(marketshare,labels=forms,shadow=True,startangle=45,explode=myexplode)
plt.title("Pie chart")
plt.legend(loc="lower center",title="List of forms")
plt.show()

```

```python
forms = ["form A","form B","form C","Form D","Form E"]
marketshare = [20,25,15,10,20]
plt.pie(marketshare,labels=forms,shadow=True,startangle=45)
plt.title("Pie chart")
plt.legend(loc="upper right",title="List of Forms")
plt.show()
```
