```python
df2 = sns.load_dataset('diamonds')
```

```python
sns.set_style("whitegrid")
sns.set_context("poster")
sns.distplot(df2['depth'],color='blue',kde=False,bins=35)
plt.show()
```

Vertical BarPlot

```python
glue = sns.load_dataset("glue")
```

```python
glue
```

```python
sns.set_style("whitegrid")
sns.set_context("notebook")
sns.barplot(data=glue , x='Model', y='Score',palette='magma',errorbar=None,color="orange")
plt.title("BAR PLOT")
plt.legend()
plt.show()
```

```python
sns.set_theme()

sns.barplot(data=glue, x='Model',y= 'Score',orient='h',palette="magma",errorbar=None)
plt.show()
```

```python
sns.countplot(data=glue, x='Score',hue='Model',palette='magma')
plt.legend()
plt.show()
```
