# **Joint Plot**

### Draws a plot of two variables with bivariate and univariate graphs, it uses a scatter plot and histogram.joint plot can also display data using kernel density estimate(KDE) and hexagons

```python
sns.set_style("dark")
tips = sns.load_dataset("tips")
tips
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips,color="orange")
```

### ADD regression line to scatter plot and KDE to histogram

```python
sns.jointplot(x="total_bill", y="tip", data=tips,color="deeppink",kind='reg')
```

### Display KDE instead of scatter plot and histogram

sns.jointplot(x="total_bill", y="tip", data=tips,color="hotpink",kind='kde')

### display hexagon instead of points in scatter plot

```python
sns.jointplot(x="total_bill", y="tip", data=tips,color="springgreen",kind='hex')
```

# BOX PLOT

```python
tips = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill",data=tips)
```

### boxplot is composed of summary of five different data points
1. Minimum
2. First Quartile or 5%
3. 3.Median (Second Quartile) or 50%
4. Third Quartile or 75%
5. Maximum
